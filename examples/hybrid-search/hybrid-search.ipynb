Combining **dense vectors** with **BM25 sparse search** using Reciprocal Rank Fusion.


```python
import sys; sys.path.insert(0, '../..')
from dotenv import load_dotenv; load_dotenv()
from hybrid_search import HybridSearchRAG

docs = [{"content": "The /users endpoint has a rate limit of 100 requests per minute."},
        {"content": "Error 429 indicates rate limit exceeded. Use exponential backoff."}]

rag = HybridSearchRAG(vector_weight=0.5, bm25_weight=0.5, use_reranker=True)
rag.index_documents(docs)
result = rag.query("What is the rate limit for /users?")
print(f"Answer: {result.answer}")
```
