Retrieval across **text, images, and audio** using vision-language models.


```python
import sys; sys.path.insert(0, '../..')
from dotenv import load_dotenv; load_dotenv()
from multimodal_rag import MultimodalRAG

rag = MultimodalRAG(vision_model="gpt-4o")
rag.index_documents([{"content": "Q4 Revenue: $5.2B, up 15% YoY. See chart."}])
rag.build_index()

result = rag.query("What was the Q4 revenue?")
print(f"Modalities: {result.modalities_used}, Answer: {result.answer}")
```
