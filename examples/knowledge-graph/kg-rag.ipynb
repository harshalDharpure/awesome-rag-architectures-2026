Combining **structured knowledge graphs** with vector search.


```python
import sys; sys.path.insert(0, '../..')
from dotenv import load_dotenv; load_dotenv()
from kg_rag import KnowledgeGraphRAG

rag = KnowledgeGraphRAG()
rag.add_entity("Apple", {"type": "Company"})
rag.add_entity("Tim Cook", {"type": "Person", "role": "CEO"})
rag.add_relationship("Tim Cook", "CEO_OF", "Apple")
rag.index_documents([{"content": "Tim Cook has led Apple since 2011."}])

result = rag.query("Who is the CEO of Apple?")
print(f"Entities: {result.entities_found}, Answer: {result.answer}")
```
